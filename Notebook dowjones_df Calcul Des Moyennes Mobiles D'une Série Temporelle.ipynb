## Moyennes mobiles

Definition : La moyenne mobile est une technique qui permet de créer une nouvelle serie temporelle à partir de notre serie temporelle originelle. Ou chaque point de données de cette serie represente la moyenne calculer sur certaine une periode donner, qui peu etre 1 jour, 10 jours, 1 mois etc.

Pourquoi utilise t-on la moyenne mobile ?
il faut savoir que la moyenne mobile entre en compte dans plusieur dommaine , la modelisation la planification et tres souvent utiliser dans les strategie de gestion et d'analyse financiere .

```python
# Moyennes mobiles sur 50 jours

dowjones_df['MA 50'] = dowjones_df['Price'].rolling(50).mean()   # MA = mooving average = moyenne mobile

# Moyennes mobiles sur 100 jours

dowjones_df['MA 100'] = dowjones_df['Price'].rolling(100).mean()   # MA = mooving average = moyenne mobile
```

## Visualisation de la serie temporelle crée

```python
# Variable qu'on veut representer

vars = ['Price', 'MA 50', 'MA 100']

# Graphique

dowjones_df[vars].plot(figsize = (8, 5));
```

On constate que nos moyennes mobiles et notre serie temporelle originelle on pratiquement la meme evolution
