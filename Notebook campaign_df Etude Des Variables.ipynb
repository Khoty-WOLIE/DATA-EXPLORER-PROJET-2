```python
click = pd.read_csv('click.csv')
click.head()
```

```python
purchase = pd.read_csv('purchase.csv')
purchase.head()
```

```python
df_final = pd.merge(campaign, click, on='client_id', how='left')
df_final.head()
```

```python
df_final = pd.merge(df_final, purchase, on='client_id', how='left')
df_final.head()
```

```python
df_final.describe(include='all')

# affiche les informations des colonnes comme, compter les valeurs dans la colonne, le frquence, la moyenne etc.
```

```python
df_final.info()

# Affiche egalement les informations comme la methode .describe()
```

```python
var_modif = ['client_id', 'campaign_id', 'canal_id', 'product_id']
for var in var_modif:
    df_final[var]=df_final[var].astype('category')


# La boucle for permet de changer les types des colonnes concernées en "catégorie" (par exemple str pour chaîne de caractères).
# La fonction astype() nécessite qu'on lui précise explicitement le type souhaité, tandis que la fonction to_numeric() 
# détecte automatiquement le type numérique, comme entier (int).*
```

```python
df_final['click_date'] = pd.to_datetime(df_final['click_date'])

# pour changer les dates qui sont en "objet ou float" en datetime
```

```python
df_final.info()
```

```python
df_final['purchase_date'] = pd.to_datetime(df_final['purchase_date'], unit='s')

# pour changer le type de la date en datetime et on precise unit="s" pour dire qu'il y a une seconde
```

```python
df_final.loc[df_final['age'] > 65, :]

# la methode .loc prend en compte la codition et la colonne, par exemple :
# dataframe.loc[dataframe['age'] > 65, :]  les : prennent en compte toute la data frame
```

```python
df_final.loc[df_final['age'] > 65, 'client_id']

# la methode .loc prend en compte la codition et la colonne, par exemple :
# dataframe.loc[dataframe['age'] > 65,'client_id' ]  la condition si la colonne "age" > 65 sur la colonne "client_id"
```

```python
df_final.loc[(df_final['age'] < 25) & (df_final['product_id'] == 8), :]

# nous avons emis deuc condition separé par et = & . on peu aussi utilisé le ou = |  ce sont les deux separateurs de condition
```

```python
df_final.loc[(df_final['age'] < 25) & (df_final['product_id'] == 8), ['campaign_id', 'client_id']]

# Apres avoir mis les deux conditions on peu selectionner plusieur colonne separé par des virgules(,)
```

```python
achats = df_final.loc[df_final['purchase_date'].notnull(), :].copy()
achats.head()


# Pour afficher les valeurs manquantes, on a les conditions suivantes :
# isnull : sélectionne l'ensemble des lignes où la variable est null. Il y a aussi isna, mais isnull est souvent préféré.
# notnull : sélectionne l'ensemble des lignes pour lesquelles les valeurs ne sont pas nulles.
# On peut stocker cette information dans une nouvelle variable, par exemple : achat = df_final.loc[df_final['purchase_date'].notnull(), :].
# .copy() : après la création d'un nouveau DataFrame avec .loc[], la méthode .copy() permet de créer deux DataFrames 
# distincts, afin d'éviter les messages d'avertissement (warnings) lors de la création de nouvelles colonnes ou lors de calculs sur celles-ci.*
```

```python
canaux = [550488, 154492, 721316, 464778, 850899]
achats.loc[achats['canal_id'].isin(canaux), :]

# la methode isin() permet d'eviter les jointures parfois. elle est beaucoup utile. elle est en rapport avec .loc[]
```

```python
achats['price TTC'] = achat['price']* 1.2
achats.head()

# ce code permet de créer une nouvelle colonne et effectuer des calcul comme multiplier ou additionner etc
# il y avait un message d'erreur comme j'ai utilisé la methode .copy() sur la creation de la data frame "achats" le
# message d'erreur a disparue
```

```python
achats['jour_achat'] = achats['purchase_date'].dt.date
achats.head()

# ce code permet de creer une nouvelle colonne et et la methone .dt.date permet d'ecrire la date sans le time(minute)
# pour la date c'est : .dt.date , on a aussi .dt.day , .dt.month etc  ; pour les chaine de caractere c'est :  .str.fonction
```

```python
achats.groupby('jour_achat').sum() [['price TTC']]

# la methode .groupby() se fixe sur une clé et utilise les fonctions de calcul soit quantitative ou qualitative
# quantitative : .sum(), .mean() ... et qualitative : .count(), .nunique etc
```
